```python
# removing unwanted colunm
covid19.drop(["Source", "Unnamed: 13", "Total Suspected Cases"], inplace = True, axis = 1)
```

```python
# renaming column
covid19.rename(columns = {"Location" : "State"}, inplace = True)
```

```python
covid19.head()
```

```python
# calculating number of active cases an creating a column of active cases
covid19["Active_Cases"] = covid19["Total Confirmed Cases"] - (covid19["Recovered"] + covid19["Deaths"]) 
```

```python
covid19.head()
```

```python
# cretion of pivot table
pivot_table = pd.pivot_table(covid19, values = ["Total Confirmed Cases", "Deaths", "Recovered"], index = "State", aggfunc = max)
```

```python
# finding out the recovery rate
pivot_table["Motality Rate"] = pivot_table["Deaths"]*100/pivot_table["Total Confirmed Cases"]
```

```python
# sorting the values base on comfirm case
pivot_table = pivot_table.sort_values(by = "Total Confirmed Cases", ascending = False)
```

```python
# plotting the pivot table
pivot_table.style.background_gradient(cmap = "cubehelix")
```
